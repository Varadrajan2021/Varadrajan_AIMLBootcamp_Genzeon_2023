```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/ML/Mall_Customers.csv')
```

```python
df=df.drop(['CustomerID'],axis=1)
```

```python

from sklearn.preprocessing import LabelEncoder

gender=df['Gender']

label_encoder = LabelEncoder()

df['Gender'] = label_encoder.fit_transform(gender)

df
```

```python
#alternative for LabelEncoding
#df['Gender']=df.gender.map({'Male':1,'Female':0})
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
#Instantiate the model
km=KMeans(n_clusters=3)# 3 centroids
clusters=km.fit_predict(df)# which datapoint into which cluster- intution
```

```python
clusters
```

```python
df['labels']=clusters
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D




fig = plt.figure(figsize=(40,30))

ax = fig.add_subplot(111, projection='3d')

ax.scatter(df.Age[df.labels == 0], df["Annual Income (k$)"][df.labels == 0], df["Spending Score (1-100)"][df.labels == 0], c='blue', s=60)

ax.scatter(df.Age[df.labels == 1], df["Annual Income (k$)"][df.labels == 1], df["Spending Score (1-100)"][df.labels == 1], c='red', s=60)

ax.scatter(df.Age[df.labels == 2], df["Annual Income (k$)"][df.labels == 2], df["Spending Score (1-100)"][df.labels == 2], c='green', s=60)


ax.view_init(30, 185)

plt.xlabel("Age")

plt.ylabel("Annual Income (k$)")

ax.set_zlabel('Spending Score (1-100)')

plt.show()
```
